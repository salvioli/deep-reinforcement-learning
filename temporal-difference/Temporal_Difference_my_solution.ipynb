### Part 1: TD Control: Sarsa

In this section, you will write your own implementation of the Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
class params:
    max_eps = 0.5
    
def generate_epsilon(index, params, method="exponential"):
    try:
        if method == "linear":
            m = (params.min_eps-params.max_eps)/params.max_i
            q = params.max_eps
            e = np.clip(m * index + q, params.min_eps, np.Infinity)
        elif method == "exponential":
            e = params.max_eps**index
    
    except AttributeError:
        raise
        
    return e
````

````python
def generate_epsilon_greedy_policy(Q, epsilon, env):
    policy = defaultdict(lambda: 0)
    for state in Q.keys():
        greedy_action = np.argmax(Q[state])
        policy[state] = greedy_action if np.random.uniform() > epsilon else env.action_space.sample()
    return policy
````

````python
def sarsa(env, num_episodes, alpha, gamma=1.0):
    # initialize action-value function (empty dictionary of arrays)
    Q = defaultdict(lambda: np.zeros(env.nA))
    # initialize performance monitor
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()   
        
        e = generate_epsilon(i_episode, params)
        state = env.reset()
        action = generate_epsilon_greedy_policy(Q, e, env)[state]
        
        done = False
        while not done:
            new_state, new_reward, done, info = env.step(action)
            new_action = generate_epsilon_greedy_policy(Q, e, env)[new_state]
            
            Q[state][action] += alpha*(new_reward + gamma*Q[new_state][new_action]-Q[state][action])
            
            state, action = new_state, new_action
        
    return Q
````

### Part 3: TD Control: Expected Sarsa

In this section, you will write your own implementation of the Expected Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
def generate_epsilon_greedy_policy_sthocastic(Q, epsilon, env):
    policy = defaultdict(lambda: np.ones(env.nA)/env.nA)
    for state in Q.keys():
        policy[state] = np.ones(env.nA)*epsilon/env.nA
        policy[state][np.argmax(Q[state])] += 1-epsilon
    return policy


def expected_sarsa(env, num_episodes, alpha, gamma=1.0):
    # initialize empty dictionary of arrays
    Q = defaultdict(lambda: np.zeros(env.nA))
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        params.max_eps = 0.005
        e = generate_epsilon(i_episode, params)
        state = env.reset()
        
        done = False
        while not done:
            pi = generate_epsilon_greedy_policy_sthocastic(Q, e, env)
            action = np.argmax(pi[state])
            
            new_state, new_reward, done, info = env.step(action)
             
            Q[state][action] += alpha*(new_reward + gamma * np.dot(pi[new_state], Q[new_state]) - Q[state][action])
            
            state = new_state
        
    return Q
````
